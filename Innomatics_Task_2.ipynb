```python
# PROBLEM 1: Unique Words in a Sentence
def count_unique_words():
    # Input from the PDF example
    sentence = "python is easy and python is powerful"

    # Logic: Split string -> Convert to set (removes duplicates)
    words_list = sentence.split()
    unique_words = set(words_list)

    print(f"Unique words count: {len(unique_words)}")
    print(f"Unique words: {unique_words}")

count_unique_words()
```

```python
# PROBLEM 2: Highest Salary from Employee Data
def find_highest_salary():
    # Input from PDF example
    employees = {
        "Ravi": 75000,
        "Anita": 68000,
        "Kiran": 72000
    }

    # Logic: Initialize variables to track max salary
    highest_salary = 0
    top_employee = ""

    for name, salary in employees.items():
        if salary > highest_salary:
            highest_salary = salary
            top_employee = name

    print(f"Highest Salary: {top_employee} - {highest_salary}")

find_highest_salary()
```

```python
# PROBLEM 3: Find Maximum and Minimum Values
def find_max_min():
    # Input from PDF example
    numbers = [45, 22, 89, 10, 66]

    # Logic: Initialize with the first element
    max_val = numbers[0]
    min_val = numbers[0]

    for num in numbers:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num

    print(f"List: {numbers}")
    print(f"Max: {max_val}")
    print(f"Min: {min_val}")

find_max_min()
```

```python
# PROBLEM 4: Count Products Above a Price Threshold
def count_expensive_products():
    # Input from PDF example
    prices = [450, 1200, 899, 1500, 300]
    threshold = 1000
    count = 0

    # Logic: Loop and increment counter
    for price in prices:
        if price > threshold:
            count += 1

    print(f"Products above {threshold}: {count}")

count_expensive_products()
```

```python
# PROBLEM 5: Calculate Attendance Percentage
def calculate_attendance():
    # Input from PDF example
    attendance = ["P", "P", "A", "P", "P"]

    total_days = len(attendance)
    present_count = 0

    # Logic: Count occurrences of "P"
    for status in attendance:
        if status == "P":
            present_count += 1

    percentage = (present_count / total_days) * 100
    print(f"Attendance Percentage: {percentage}")

calculate_attendance()
```

```python
# PROBLEM 6: Remove Duplicate Phone Numbers
def clean_phone_numbers():
    # Input from PDF example
    phone_numbers = [9876543210, 9123456789, 9876543210]

    # Logic: Typecast list to set
    unique_numbers = set(phone_numbers)

    print(f"Unique phone numbers: {unique_numbers}")

clean_phone_numbers()
```

```python
# PROBLEM 7: Count Character Frequency
def char_frequency():
    input_str = "python"  # Based on expected output {'p': 1, etc...}
    freq_dict = {}

    # Logic: Loop through string, update dictionary
    for char in input_str:
        if char in freq_dict:
            freq_dict[char] += 1
        else:
            freq_dict[char] = 1

    print(freq_dict)

char_frequency()
```

```python
# PROBLEM 8: Convert List to Tuple
def list_to_tuple():
    # Logic: Use tuple() constructor
    # Implicit input based on output (10, 20, 30)
    my_list = [10, 20, 30]
    my_tuple = tuple(my_list)

    print(f"Tuple: {my_tuple}")

list_to_tuple()
```

```python
# PROBLEM 9: Check if a Key Exists in a Dictionary
def check_key_exists():
    # Implicit input structure
    employees = {"Ravi": 75000, "Anita": 68000}
    key_to_check = "Ravi"

    # Logic: direct check
    if key_to_check in employees:
        print("Employee exists")
    else:
        print("Employee does not exist")

check_key_exists()
```

```python
# PROBLEM 10: Calculate Average Marks
def calculate_average():
    # Implicit input based on output 72.5
    # (Example: 70 + 75 = 145 / 2 = 72.5)
    marks = [70, 75, 80, 65]

    total_marks = sum(marks)
    count = len(marks)

    average = total_marks / count
    print(f"Average Marks: {average}")

calculate_average()
```
