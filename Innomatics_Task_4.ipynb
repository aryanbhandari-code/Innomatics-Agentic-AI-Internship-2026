```python
# PROBLEM 1: Mobile Recharge Validation System
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount (or 0 to exit): "))

            if amount == 0:
                print("Exiting system.")
                break
            elif amount < 50:
                print("Amount must be >= ₹50. Please try again.")
            elif amount in valid_plans:
                print(f"Success! Recharge of ₹{amount} is processing.")
                break
            else:
                print(f"Invalid plan. Available plans are {valid_plans}. Try again.")
        except ValueError:
            print("Invalid input. Please enter numbers only.")

validate_recharge()
```

```python
# PROBLEM 2: Inventory Reorder Alert System
def check_inventory(inventory):
    print("--- Inventory Report ---")
    for product, stock in inventory.items():
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock: {stock})")
        else:
            print(f"{product}: Stock OK (Stock: {stock})")

# Test Data
warehouse_stock = {"Laptops": 12, "Mice": 45, "Keyboards": 10, "Monitors": 20}
check_inventory(warehouse_stock)
```

```python
# PROBLEM 3: Student Result Processing System
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        print(f"Average: {average:.2f} -> Pass")
    else:
        print(f"Average: {average:.2f} -> Fail")

# Test Data
student_marks = [45, 60, 55, 70, 48]
process_result(student_marks)
```

```python
# PROBLEM 4: Cab Fare Estimator with Retry
def estimate_cab_fare():
    base_fare = 50
    per_km_rate = 12

    while True:
        try:
            distance = float(input("\nEnter distance in km: "))
            peak_hour = input("Is it peak hour? (yes/no): ").strip().lower()

            # Base calculation
            total_fare = base_fare + (per_km_rate * distance)

            # Peak hour surcharge
            if peak_hour == 'yes':
                total_fare += total_fare * 0.25  # Add 25%

            print(f"Estimated Cab Fare: ₹{total_fare:.2f}")

            # Retry mechanism
            retry = input("Calculate another fare? (yes/no): ").strip().lower()
            if retry != 'yes':
                print("Thank you for using the Cab Estimator.")
                break
        except ValueError:
            print("Invalid distance entered. Please try again.")


estimate_cab_fare()
```

```python
# PROBLEM 5: Employee Attendance Eligibility
def check_attendance_eligibility(attendance):
    present_days = 0
    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    if percentage >= 75:
        return f"Eligible (Attendance: {percentage:.1f}%)"
    else:
        return f"Not Eligible (Attendance: {percentage:.1f}%)"

# Test Data
emp_attendance = ["P", "P", "A", "P", "P", "A", "P", "P", "P", "P"]
print(check_attendance_eligibility(emp_attendance))
```

```python
# PROBLEM 6: Password Strength Checker
def check_password_strength(password):
    if len(password) < 8:
        return "Weak: Password must be at least 8 characters long."

    has_digit = False
    has_special = False
    valid_specials = "@#$"

    for char in password:
        if char.isdigit():
            has_digit = True
        elif char in valid_specials:
            has_special = True

    if has_digit and has_special:
        return "Strong: Password meets all criteria."
    else:
        return "Weak: Password must contain at least one digit and one special character (@, #, $)."

# Test Data
print(check_password_strength("Aryan123"))
print(check_password_strength("Ary@n123"))
print(check_password_strength("Pass@"))
```
