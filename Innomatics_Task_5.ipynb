```python
# PROBLEM 1: Social Media - Post Engagement Analyzer
def analyze_engagement(daily_likes):
    total_likes = 0
    # [cite_start]Loop to calculate total as per the hint [cite: 430]
    for likes in daily_likes:
        total_likes += likes

    print(f"Total Likes: {total_likes}")
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")

# Test Data
post_likes = [200, 350, 400, 100]
analyze_engagement(post_likes)
```

```python
# PROBLEM 2: Healthcare - Medicine Stock Alert System
def check_medicine_stock(stock):
    print(f"Medicine Stock: {stock}")
    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")

# Test Data
check_medicine_stock(6)
```

```python
# PROBLEM 3: Agriculture - Rainfall Adequacy Checker
def analyze_rainfall(rainfall_data, required_level):
    total_rain = 0
    for rain in rainfall_data:
        total_rain += rain

    average_rainfall = total_rain / len(rainfall_data)

    print(f"Average Rainfall: {average_rainfall:.0f}")
    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

# Test Data
rain_records = [65, 80, 75, 68]
analyze_rainfall(rain_records, 70)
```

```python
# PROBLEM 4: Social Media - Duplicate Account Detection
def detect_duplicate_accounts(usernames):
    # A set automatically removes duplicates. If lengths differ, duplicates exist.
    unique_users = set(usernames)

    if len(usernames) > len(unique_users):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

# Test Data
users = ["aryan123", "jain_coder", "aryan123", "tech_guru"]
detect_duplicate_accounts(users)
```

```python
# PROBLEM 5: Healthcare - Appointment Eligibility Checker
def check_appointment_eligibility(age):
    print(f"Patient Age: {age}")
    if age >= 18:
        print("Eligibility Status: Eligible")
    else:
        print("Eligibility Status: Not Eligible")

# Test Data
check_appointment_eligibility(21)
```

```python
# PROBLEM 6: Agriculture - Premium Crop Price Filter
def filter_premium_crops(prices):
    premium_crops = []
    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print(f"Premium Crops: {premium_crops}")

# Test Data
crop_prices = [1500, 2500, 1800, 3200, 1950]
filter_premium_crops(crop_prices)
```

```python
# PROBLEM 7: System Monitoring - Application Health Checker
def application_health_checker(errors):
    print(f"Error Count: {errors}")
    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")

# Test Data
application_health_checker(7)
```

```python
# PROBLEM 8: Banking - Daily Transaction Limit Checker
def validate_transaction(amount):
    print(f"Transaction Amount: {amount}")
    if amount <= 50000:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")

# Test Data
validate_transaction(60000)
```

```python
# PROBLEM 9: E-Learning - Student Attendance Eligibility System
def check_exam_eligibility(attendance_list):
    present_count = 0
    for status in attendance_list:
        if status == "P":
            present_count += 1

    percentage = (present_count / len(attendance_list)) * 100

    print(f"Attendance Percentage: {percentage:.1f}")
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

# Test Data
attendance = ["P", "P", "P", "A", "P"]
check_exam_eligibility(attendance)
```

```python
# PROBLEM 10: Smart Electricity Bill Analyzer
def analyze_electricity_bill(units):
    bill = 0

    # Slab Calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

# Test Data
units_consumed = 250
total_bill, usage_status = analyze_electricity_bill(units_consumed)
print(f"Units Consumed: {units_consumed}")
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Status: {usage_status}")
```
