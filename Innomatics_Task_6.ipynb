```python
# PROBLEM 1: Smart Parking Lot Management System
def manage_parking(capacity, logs):
    current_parked = 0
    peak_usage = 0

    for action in logs:
        if action == "IN":
            current_parked += 1
            if current_parked > peak_usage:
                peak_usage = current_parked
        elif action == "OUT" and current_parked > 0:
            current_parked -= 1

    print(f"Currently Parked Vehicles: {current_parked}")
    print(f"Peak Parking Usage: {peak_usage}")

    if current_parked > capacity:
        print("Parking Status: Alert - Capacity Exceeded!")
    else:
         print("Parking Status: Available")

# Test Data based on Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking(capacity, vehicle_logs)
```

```python
# PROBLEM 2: Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic, weather):
    # Base calculation
    time = distance * 5

    # Traffic conditions
    if traffic.lower() == "high":
        time += 10
    elif traffic.lower() == "moderate":
        time += 5

    # Weather conditions
    if weather.lower() == "rainy":
        time += 5
    elif weather.lower() == "stormy":
        time += 15

    print(f"Estimated Delivery Time: {time} minutes")

# Test Data based on Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

```python
# PROBLEM 3: Movie Theatre Seat Occupancy Analyzer
def analyze_occupancy(total_seats, booked_seats_list):
    booked_count = sum(booked_seats_list)
    occupancy_percentage = (booked_count / total_seats) * 100

    print(f"Occupancy: {occupancy_percentage:.0f}%")

    if occupancy_percentage == 100:
         print("Show Status: Housefull")
         print("Suggestion: Open additional shows immediately.")
    elif occupancy_percentage >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional shows.")
    else:
        print("Show Status: Available")

# Test Data based on Sample Input
# Simulating 150 entries of '1' for booked seats
booked_list = [1] * 150
analyze_occupancy(200, booked_list)
```

```python
# PROBLEM 4: Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    total_load = sum(cpu_readings)
    average_load = total_load / len(cpu_readings)

    print(f"Average CPU Load: {average_load:.0f}%")

    if average_load < 50:
        print("Server Status: Normal")
    elif average_load <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

# Test Data based on Sample Input
readings = [45, 60, 70, 85, 90]
classify_server_load(readings)
```

```python
# PROBLEM 5: Smart Classroom Resource Usage Monitor
def monitor_resources(usage_data):
    overuse_threshold = 8
    overused_items = []

    for resource, hours in usage_data.items():
        if hours > overuse_threshold:
            overused_items.append(resource)

    if overused_items:
        print(f"Overused Resources: {', '.join(overused_items)}")
        print("Energy Alert: Yes")
    else:
        print("All resources within normal usage limits.")
        print("Energy Alert: No")

# Test Data based on Sample Input
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
monitor_resources(resources)
```

```python
# PROBLEM 6: Online Event Registration Capacity Controller
def manage_registrations(capacity, registrations):
    confirmed = 0
    waitlisted = 0
    status = ""

    if registrations <= capacity:
        confirmed = registrations
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"

    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")

# Test Data based on Sample Input
manage_registrations(100, 105)
```
