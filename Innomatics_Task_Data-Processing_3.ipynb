```python
# PROBLEM 1: Employee Performance Bonus Eligibility
def performance_bonus():
    # Input Data
    employees = {
        "Ravi": 92,
        "Anita": 88,
        "Kiran": 92,
        "Suresh": 85
    }

    # Logic:
    # 1. Find the maximum score from the values.
    # 2. Iterate through the dictionary to find all keys (names) matching that score.

    max_score = max(employees.values())

    # List comprehension to get all names with the max_score
    top_performers = [name for name, score in employees.items() if score == max_score]

    print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")

performance_bonus()
```

```python
# PROBLEM 2: Search Query Keyword Analysis
import string

def analyze_search_query():
    # Input Data
    query = "Buy mobile phone buy phone online"

    # Logic:
    # 1. Convert to lowercase.
    # 2. Remove punctuation (though input has none, it's a requirement).
    # 3. Split into words and count frequencies.

    clean_query = query.lower().translate(str.maketrans('', '', string.punctuation))
    words = clean_query.split()

    word_counts = {}
    for word in words:
        word_counts[word] = word_counts.get(word, 0) + 1

    # Filter for words appearing more than once
    # Note: The expected output format is specific {'buy': 2, 'phone': 2}
    duplicates = {word: count for word, count in word_counts.items() if count > 1}

    print(duplicates)

analyze_search_query()
```

```python
# PROBLEM 3: Sensor Data Validation
def validate_sensor_data():
    # Input Data
    sensor_readings = [3, 4, 7, 8, 10, 12, 5]

    # Logic:
    # 1. Enumerate gives us both index (hour) and value.
    # 2. Check if value % 2 == 0 (Even).

    valid_readings = []

    for hour, reading in enumerate(sensor_readings):
        if reading % 2 == 0:
            valid_readings.append((hour, reading))

    print("Valid Sensor Readings (Hour, Value):")
    print(valid_readings)

validate_sensor_data()
```

```python
# PROBLEM 4: Email Domain Usage Analysis
def analyze_email_domains():
    # Input Data
    emails = [
        "ravi@gmail.com",
        "anita@yahoo.com",
        "kiran@gmail.com",
        "suresh@gmail.com",
        "meena@yahoo.com"
    ]

    # Logic:
    # 1. Split email by '@' to get the domain.
    # 2. Count occurrences.
    # 3. Calculate percentage = (count / total) * 100.

    domain_counts = {}
    total_emails = len(emails)

    for email in emails:
        domain = email.split('@')[1]
        domain_counts[domain] = domain_counts.get(domain, 0) + 1

    for domain, count in domain_counts.items():
        percentage = (count / total_emails) * 100
        print(f"{domain}: {percentage:.0f}%")

analyze_email_domains()
```

```python
# PROBLEM 5: Sales Spike Detection
def detect_sales_spike():
    # Input Data
    sales = [1200, 1500, 900, 2200, 1400, 3000]
    average_sales = sum(sales) / len(sales)
    threshold = average_sales * 1.3

    for day_index, sale_value in enumerate(sales):
        if sale_value > threshold: # Strict math check
            print(f"Day {day_index + 1}: {sale_value}")
        # If strict math excludes 2200, it's correct based on the prompt text "more than".

detect_sales_spike()
```

```python
# PROBLEM 6: Duplicate User ID Detection
def detect_duplicate_users():
    # Input Data
    user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

    # Logic:
    # 1. Count frequency of all users.
    # 2. Display only those with count > 1.

    id_counts = {}
    for user in user_ids:
        id_counts[user] = id_counts.get(user, 0) + 1

    for user, count in id_counts.items():
        if count > 1:
            print(f"{user} {count} times")

detect_duplicate_users()
```
